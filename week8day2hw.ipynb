## Homework

Design a flexible OOP system to describe characters in a RPG game

There are 4 Unique Characters which use Different Types of Attacking mechanisms and Defending mechanisms.

- Orges - fights with club and defends with shield
- Peons - fights with club and defends with shield
- Knights - fights with a Sword and defends with Armor
- Archer - fights with Bow and Arrow and has only is tunic to protect them

- All Characters can collect gold and this will always be the same for every new type of character

At the end of this each character should be able to attack and defend and collect gold. If needed any character should be able to change his fighting style, say if a knight losing his sword, but finds a club he should be able to change his fighting mechanism to use the club

<b>Hint: </b> To complete this assignment you will need multiple classes and interfaces and an Abstract class

You can use our [Final Rubber Ducky Walkthrough](#rubber) example as a guide 

```typescript
//Solution

interface Fightable{
    fight():void
}

interface Defendable{
    defend():void
}

class Fighter implements Fightable{

    
    constructor(private name:string , private weapon:string){
    }
    
    fight(){
        console.log(`${this.name} fights with ${this.weapon}.`)
    }
}

class Defender implements Defendable{
    
    constructor( private name:string,private defence:string){ }
    
    defend(){
        console.log(`${this.name} defends with ${this.defence}.`)
    }

}

abstract class rpgClass implements Fightable, Defendable{

    //Properties
    protected gold$:number=0;
    
    //Abilities
    protected fightAbility$: Fightable;
    protected defenseAbility$: Defendable;

    //Attack setting

    constructor(protected name$:string){ }
    
    set fightAbility(fa:Fightable){
        this.fightAbility$=fa
    }

    fight(){
        this.fightAbility$.fight()
    }

    getWeapon(weapon:string){
        this.fightAbility$= new Fighter(this.name$,weapon)
    }

    //Defense setting

    set defenseAbility(da:Defendable){
        this.defenseAbility$=da
    }

    defend(){
        //if there is a defense ability use it
        if(this.defenseAbility$){
            this.defenseAbility$.defend()                    
        }else{
            console.log(`${this.name$} has only is tunic to protect them.`)
        }
    }

    collectGold(gold:number){
        this.gold$ += gold
        console.log(this.gold)
    }

    get gold(){
        return `${this.name$} now has ${this.gold$} gold`
    }
    
    getDefence(defence:string){
        this.defenseAbility$= new Defender(this.name$,defence)
    }
}

// Ogre Class
class Orge extends rpgClass{

    constructor(name$:string){
        super(name$)
        this.fightAbility$ = new Fighter(name$,"club")
        this.defenseAbility$ = new Defender(name$,"shield")
    }
}

// Archer Class
class Archer extends rpgClass{

    constructor(name$:string){
        super(name$)
        this.fightAbility$ = new Fighter(name$,"bow and arrow")
    }
}

// Knight Class
class Knight extends rpgClass{
    constructor(name$:string){
        super(name$)
        this.fightAbility$ = new Fighter(name$,"sword")
        this.defenseAbility$ = new Defender(name$,"Armor")
    }
}

// Peon Class
class Peon extends rpgClass{
    constructor(name$:string){
        super(name$)
        this.fightAbility$ = new Fighter(name$,"club")
        this.defenseAbility$ = new Defender(name$,"shield")
    }
}


// Instances

let shrek:Orge = new Orge("Shrek")

shrek.fight()
shrek.getWeapon("sword")
shrek.fight()
shrek.defend()
shrek.collectGold(20)
shrek.collectGold(30)
console.log("\n")

let altan:Peon = new Peon("Altan")
altan.fight()
altan.defend()
console.log("\n")


let princeCharming:Knight = new Knight("Prince Charming")
princeCharming.fight()
princeCharming.defend()
console.log("\n")

let legolas:Archer = new Archer("Legolas")
legolas.fight()
legolas.getWeapon("sword")
legolas.fight()
legolas.defend()
```
